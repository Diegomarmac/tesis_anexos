# CENAPI OVERVIEW

```python
from IPython.display import display

import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('data/cenapi.csv')

df.columns = df.columns.str.lower().str.replace(' ', '_')
string_columns = list(df.dtypes[df.dtypes == 'object'].index)
for col in string_columns:
    df[col] = df[col].str.lower().str.replace(' ', '_')
    
df.head()
```

```python
len(df)
```

## Cantidad de desaparecidos por genero

```python
df.x12.value_counts()
```

```python
plt.figure(figsize=(8, 8))

font1 = {'family':'sans-serif','color':'black','size':20}
font2 = {'family':'sans-serif','color':'black','size':15}

plt.ylabel('Frequency',fontdict=font2)
plt.xlabel('Gender',fontdict=font2)
plt.title('Distribution of gender', fontdict=font1)

sns.histplot(df.x12)
plt.show()
```

## Estados donde fueron declarados desaparecidos

```python
df.x06.value_counts() # esto es el estado de origen de los desaparecidos
```

```python
plt.figure(figsize=(25, 10))

font1 = {'family':'sans-serif','color':'black','size':20}
font2 = {'family':'sans-serif','color':'black','size':15}

plt.ylabel('Frequency',fontdict=font2)
plt.xlabel('State', fontdict=font2)
plt.title('Distribution of Missing', fontdict=font1)


sns.histplot(df.x06,kde=True)
plt.show()
```
