# Analisis de los cluster de la submuestra de 13 a 18

```python
from IPython.display import display

import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('13-17_clustered.csv')
df.head()
```

```python
df["cluster"].value_counts()
```

```python
plt.figure(figsize=(8, 8))

sns.set_palette("coolwarm")
sns.set_style("ticks")
sns.set_context("talk")

plt.ylabel('Frequency')
plt.xlabel('Cluster')
plt.title('Distribution of Cluster,13 to 18 age,all Mexico')

ax = sns.histplot(df.cluster, discrete='True')
ax.bar_label(ax.containers[0])

plt.show()
```

```python
cero = df[df["cluster"] == 0].copy()
uno = df[df["cluster"] == 1].copy()
dos = df[df["cluster"] == 2].copy()
```

```python
fig, axs = plt.subplots(1, 3,figsize=(25,8))

fig.suptitle('Estatus de localizacin x Clulster')

plotUno = sns.histplot(cero.vivo_o_muerto,ax=axs[0])
plotUno.bar_label(plotUno.containers[0])
axs[0].set_title('Cluster 0')

plotDos = sns.histplot(uno.vivo_o_muerto,ax=axs[1])
plotDos.bar_label(plotDos.containers[0])
axs[1].set_title('Cluster 1')

plotTres = sns.histplot(dos.vivo_o_muerto,ax=axs[2])
plotTres.bar_label(plotTres.containers[0])
axs[2].set_title('Cluster 2')
```

```python
plt.figure(figsize=(20,10))

sns.set_style("ticks")
sns.set_context("talk")

plot_cero = sns.histplot(cero.estado, label='Cluster 0', alpha=0.5,linewidth=.8,discrete="true",color="#fe218b")
plot_uno = sns.histplot(uno.estado,label='Cluster 1', alpha=0.5,linewidth=.8,discrete="true",color="#fed700")
plot_dos = sns.histplot(dos.estado, label='Cluster 2', alpha=0.8,linewidth=.8,discrete="true", color="#21b0fe")

plot_cero.set_xticklabels(plot_cero.get_xticklabels(), rotation = 60)
plot_uno.set_xticklabels(plot_uno.get_xticklabels(), rotation = 60)
plot_dos.set_xticklabels(plot_dos.get_xticklabels(), rotation = 60)

plt.legend()

plt.ylabel('Frecuencia')
plt.xlabel('Estado')
plt.title('Estado x Cluster')

plt.show()
```
