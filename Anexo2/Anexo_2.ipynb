# File to clean the original database to work with it

```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data/given_by_s/cenapi.csv')

df.columns = ["REGISTRO NACIONAL","FECHA REPORTE","HORA REPORTE","FECHA EVENTO","HORA EVENTO","ESTADO","CLAVE ESTADO","MUNICIPIO","CLAVE MUNICIPIO","NACIONALIDAD",
"ESTATUS MIGRATORIO","SEXO","FECHA DE NACIMIENTO","EDAD","SEÑAS PARTICULARES","SEnAS PARTICULARES","MEDIA FILIACIoN","ENFERMEDAD","DISCAPACIDAD MENTAL","TIPO DE DISCAPACIDAD MENTAL",
"DESCRIPCIoN DE VESTIMENTA","CONTABA CON APARATOS DE COMUNICACIoN","OCUPACIoN","RELACIoN CON GRUPOS DELICTIVOS","FECHA DE ULTIMO AVISTAMIENTO","HORA DE ULTIMO AVISTAMIENTO",
"TIPO DE VEHiCULO","MARCA DE VEHiCULO","AnO DEL VEHiCULO","DISCAPACIDAD FiSICA","PROCEDENCIA","PAiS DE ORiGEN","ENTIDAD","ETNIA","ANTEOJOS","FECHA DE LOCALIZACIoN","HORA DE LOCALIZACIoN",
"VIVO O MUERTO","POSIBLE CAUSA DESAPARICIoN","CONDICIoN ENCONTRADO","ESTADO (LOCALIZADO)","CLAVE ESTADO (LOCALIZADO)","MUNICIPIO (LOCALIZADO)","CLAVE MUNICIPIO (LOCALIZADO)","ENTE QUE LOCALIZA",
"FECHA DE INGRESO","HORA DE INGRESO","FECHA PROBABLE DE FALLECIMIENTO","CAUSAS DE FALLECIMIENTO","PESO","ESTATURA","FORMA DE LA CARA","COLOR DE OJOS","TIPO DE OJOS","TEZ","FORMA DEL CABELLO","LARGO DEL CABELLO",
"COLOR DEL CABELLO","TAMAnO DE OREJAS","FORMA DEL MENToN","TAMAnO DE LA NARIZ","TIPO DE NARIZ","GROSOR DE LABIOS","TIPO DE FRENTE","CEJAS","TIPO DE CEJAS","TAMAnO DE LA BOCA","BARBA","BIGOTE",
"SUBMARCA","TIPO DENUNCIA","TIPO DE EVENTO","TIPO CAUSAL","CLASIFICACION CAUSAL","CAUSAL"]

df.head()
```

```python
for element in ["FECHA REPORTE",'FECHA EVENTO', 'FECHA DE ULTIMO AVISTAMIENTO', 'FECHA DE INGRESO', 'FECHA PROBABLE DE FALLECIMIENTO', "FECHA DE LOCALIZACIoN", "FECHA DE NACIMIENTO", "FECHA PROBABLE DE FALLECIMIENTO"]:
    df[element] = df[element].str.replace(r' 0:00$', '')

df[["FECHA REPORTE",'FECHA EVENTO', 'FECHA DE ULTIMO AVISTAMIENTO', 'FECHA DE INGRESO', 'FECHA PROBABLE DE FALLECIMIENTO', "FECHA DE LOCALIZACIoN", "FECHA DE NACIMIENTO", "FECHA PROBABLE DE FALLECIMIENTO"]].head()
```

## NOTA inconsistencias en los datos: en la db original los siguientes datos tienen como año de evento 1900 pero el reporte es después del año 2000, con lo cual todos los siguientes datos serán tratados como NaT en python, con lo cual podremos observar datos erroneos

```python
df[df["FECHA EVENTO"] == "01/01/00"]
```

```python
df["FECHA REPORTE"] = df["FECHA REPORTE"] + " " + df["HORA REPORTE"]
df["FECHA EVENTO"] = df["FECHA EVENTO"] + " " + df["HORA EVENTO"]
df["FECHA DE ULTIMO AVISTAMIENTO"] = df["FECHA DE ULTIMO AVISTAMIENTO"] + " " + df["HORA DE ULTIMO AVISTAMIENTO"]
df["FECHA DE LOCALIZACIoN"] = df["FECHA DE LOCALIZACIoN"] + " " + df["HORA DE LOCALIZACIoN"]
df["FECHA DE INGRESO"] = df["FECHA DE INGRESO"] + " " + df["HORA DE INGRESO"]

df[["FECHA REPORTE",'FECHA EVENTO', 'FECHA DE ULTIMO AVISTAMIENTO', 'FECHA DE INGRESO', "FECHA DE LOCALIZACIoN"]].head()
```

```python
for element in ["FECHA REPORTE",'FECHA EVENTO', 'FECHA DE ULTIMO AVISTAMIENTO', 'FECHA DE INGRESO', "FECHA DE LOCALIZACIoN",'FECHA PROBABLE DE FALLECIMIENTO']:
    df[element] = pd.to_datetime(df[element],errors="coerce") # el errors coerce, tomará todos los errores en fechas como NaT, con lo que vemos inconcistencias

df[["FECHA REPORTE",'FECHA EVENTO', 'FECHA DE ULTIMO AVISTAMIENTO', 'FECHA DE INGRESO', 'FECHA PROBABLE DE FALLECIMIENTO', "FECHA DE LOCALIZACIoN"]].head()
```

```python
for element in ["HORA REPORTE","HORA EVENTO","HORA DE ULTIMO AVISTAMIENTO","HORA DE LOCALIZACIoN","HORA DE INGRESO"]:
    del df[element]
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
string_columns = list(df.dtypes[df.dtypes == 'object'].index)
for col in string_columns:
    df[col] = df[col].str.lower().str.replace(' ', '_')
    
df.head()
```
